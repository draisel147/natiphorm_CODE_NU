```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
image = cv2.imread('/content/drive/MyDrive/Image Processing/Pic/man_moon.png', 0)
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
total_pixels = image.shape[0] * image.shape[1]
max_frequency = max(hist)
if max_frequency / total_pixels > 0.7:
   print("ภาพสว่าง")
elif max_frequency / total_pixels < 0.2:
   print("ภาพมืด")
else:
   print("Low contrast")
cv2_imshow(image)
```

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

image =cv2.imread('/content/drive/MyDrive/Image Processing/Pic/lena.png', 0)

row,col = image.shape

img2 = np.zeros((row,col),dtype='uint8')
threshold=int(input("Enter Threshold :"))

for i in range(row):
  for j in range(col):
    if image[i,j]<threshold:
      img2[i,j]=0
    else:
      img2[i,j]=255

cv2_imshow(image)
cv2_imshow(img2)

```

```python
import cv2
import numpy as np
img1 = cv2.imread('/content/drive/MyDrive/Image Processing/Pic/chest_xray1.jpeg', cv2.IMREAD_GRAYSCALE)

img2 = cv2.imread('/content/drive/MyDrive/Image Processing/Pic/chest_xray2.jpeg', cv2.IMREAD_GRAYSCALE)

hist1 = cv2.calcHist([img1], [0], None, [256], [0, 256])
hist2 = cv2.calcHist([img2], [0], None, [256], [0, 256])

diff = np.sum(np.abs(hist1 - hist2))

if diff == 0:
  print("ภาพเหมือนกัน")
else:
  print("ภาพต่างกัน")
print("มีผลต่างเท่ากับ =", diff)
```
